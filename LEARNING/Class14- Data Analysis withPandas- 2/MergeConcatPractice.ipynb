```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'key':['a','b','c','d','e'],
                   'A1': range(5),
                   'B1': range(5,10)})
df2 = pd.DataFrame({'key':['a','b','c'],
                   'A2': range(3),
                   'B2': range(3,6)})
df2
```

```python
df4 = pd.merge(df1,df2,how="left",on="key")
df4
```

```python
pd.merge(df1,df2,how="inner",on="key")
```

```python
pd.concat([df1,df2])
```
