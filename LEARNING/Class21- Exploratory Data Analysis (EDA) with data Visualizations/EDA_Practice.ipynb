```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('taitanic.csv')
```

```python
df.head()
```

**How many Servived?**

```python
s1 = df['Survived'].value_counts()
```

```python
s1.plot.pie(autopct='%.2f%%',shadow=False)
```

```python
datalebel = sns.countplot(x=df['Survived'])
for i in datalebel.containers:
    datalebel.bar_label(i)
```

```python
f,axe = plt.subplots(1,2,figsize=(12,5))
df ['Survived'].value_counts().plot.pie(autopct='%.2f%%',ax=axe[0],shadow=True)
axe[0].set_title('Survived')

datalabel = sns.countplot(x=df['Survived'],ax=axe[1],hue="Sex",data=df)
axe[1].set_title('Survived')

for i in datalabel.containers:
    datalabel.bar_label(i,)

plt.show()
```
