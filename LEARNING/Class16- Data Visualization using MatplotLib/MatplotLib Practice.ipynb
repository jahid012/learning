```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(11)
y = x * 2
print(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-*')
```

```python
plt.plot(x,y,'r-*')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Figure/plot Title')
plt.xticks(range(9,11))
plt.yticks(range(0,101,10))
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r*-')
plt.subplot(1,3,2)
plt.plot(x,y,'g*-')
plt.subplot(1,3,3)
plt.plot(x,y,'b--')
plt.tight_layout()
```

```python
fig1 = plt.figure(figsize=(8,5))

axes = fig1.add_axes([0,0,1,1])

axes.plot(x,y,'r-*')

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("Figure Title")

axes.set_yticks(range(0,101,10))
axes.set_xticks(range(0,11));
fig1.savefig("graphsave.pdf",dpi=200)
```

```python
fig, axes = plt.subplots(2,2)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(2,2,sharex=True,sharey=True,figsize=(8,5))

axes[0][0].plot(x,y,'r-*')
axes[0][0].set_xlabel('X Label')
axes[0][0].set_title('Figure Title')

axes[1,1].scatter(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label="x**2")
axes.plot(x,x**3,label="x**3")

axes.legend(loc=2)
plt.show()
```
